```python
'''1. Collect data from the Franfurt Stock Exchange
, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).'''

import requests

url = "https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X/data.json"
params = {"start_date": "2017-01-01", "end_date": "2017-12-31"}
headers = {"X-API-KEY": API_KEY}

response = requests.get(url, params=params, headers=headers)

if response.status_code == 200:
    with open("nasdaq_data.json", "wb") as f:
        f.write(response.content)
        print("Data downloaded successfully!")
else:
    print("Error downloading data:", response.status_code)

```

```python
#2.Convert the returned JSON object into a Python dictionary

import json #import json library to process Json file in python
file = open('nasdaq_data.json', 'r') # Open the file in read format
data_dict = json.load(file)  # Using Json module, load, convert the file to json object
print(data_dict)
```

```python
print(json.dumps(data_dict, indent=4)) #print json data in a readable format using dumps method in json library
```

```python
for key, value in data_dict.items():
    print(key, value)
```

```python
print(data_dict["dataset_data"]["data"])
```

```python
#JSON file has column names and values in different lists.
#pull the column names and values to 2 separate lists from json file
column_names = data_dict["dataset_data"]["column_names"]
values = data_dict["dataset_data"]["data"]
print(column_names)
print(values)
```

```python
# Create a list of dictionaries from the column names and values
data_list = []
for row in values:
    row_dict = dict(zip(column_names, row)) # combine 2 lists using zip and create a dictionary
    data_list.append(row_dict)

# Now you can process the list of dictionaries as desired
for row_dict in data_list:
    if row_dict["Open"] is None:
        print(row_dict)
```

```python
#3.Calculate what the highest and lowest opening prices were for the stock in this period.
highest_opening_price = max([d["Open"] for d in data_list if d["Open"] is not None])
print(highest_opening_price)
```

```python
lowest_opening_price = min([d["Open"] for d in data_list if d["Open"] is not None])
print(lowest_opening_price)
```

```python
#4.What was the largest change in any one day (based on High and Low price)?
maximum_change = max([d["Change"] for d in data_list if d["Change"] is not None])
print(maximum_change)
```

```python
maximum_change_in_one_day = max([d["High"]-d["Low"] for d in data_list if d["Low"] is not None and d["High"] is not None])
print(maximum_change_in_one_day)
```

```python
maximum_change_in_one_day_date = [d["Date"] for d in data_list if (d["High"] - d["Low"]) == maximum_change_in_one_day]
print(maximum_change_in_one_day_date)
```

```python
sorted_data = sorted(data_list, key=lambda x: x["Date"], reverse=False)
print(sorted_data)
```

```python
#5.What was the largest change between any two days (based on Closing Price)?
max_diff=0
for i in range(len(sorted_data)-1):
    diff=sorted_data[i+1]["Close"]-sorted_data[i]["Close"]
    if diff>max_diff:
        max_diff=diff
        the_date=sorted_data[i+1]["Date"]
print(max_diff)
print(the_date)
```

```python
#6.What was the average daily trading volume during this year?
total=0
for i in range(len(sorted_data)):
    sum_value=sorted_data[i]["Traded Volume"]
    total=total+sum_value
avg=total/len(sorted_data)
print(avg)
```
